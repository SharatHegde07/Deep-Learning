```python
import cv2
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.applications.imagenet_utils import preprocess_input
import numpy as np
```

```python
model = load_model('Face_mask.h5')
```

```python
cam = cv2.VideoCapture(0)
face_class = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
while True:
    _,img = cam.read()
    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_class.detectMultiScale(gray, 1.3, 5)
    for (x, y, w ,h) in faces:
        
        face = img[y:y+h, x:x+w]
        face = cv2.resize(face, (224, 224))
        face = img_to_array(face)
        face = np.expand_dims(face, axis = 0)
        face = preprocess_input(face)
        text = " "
        color = (0, 0, 0)
        
        (mask, without_mask) = model.predict(face)[0]
        if mask > without_mask:
            text = "Mask Detected"
            color = (0, 255, 0)
        else:
            text = "No Mask!"
            color = ( 0, 0, 255)
        print(text)
         
        cv2.rectangle(img, (x, y), (x+w, y+h), color, 2)
        cv2.putText(img, text, (x, y-15), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)
    
    cv2.imshow('Mask Detector', img)
    if cv2.waitKey(1) == 13:
        break
cam.release()
cv2.destroyAllWindows()
```
